## 0.3 Setting up a virtual environment

A virtual environment is an isolated Python setup for your project. It keeps your packages separate from other projects and the system Python. This prevents version conflicts and makes your setup reproducible.

**Create the virtual environment:**
```bash
python3 -m venv venv
```

This creates a folder called `venv` inside your project folder.

**Activate the virtual environment:**

Windows (Command Prompt):
```bash
venv\Scripts\activate
```

Windows (PowerShell):
```bash
venv\Scripts\Activate.ps1
```

Mac/Linux:
```bash
source venv/bin/activate
```

When activated, you'll see `(venv)` at the start of your terminal prompt. This tells you you're working inside the virtual environment.

**Important:** You need to activate the virtual environment every time you open a new terminal to work on this project.

> **Escape hatch:**
If the virtual environment gives you persistent trouble, you can skip it and install packages globally with `pip install jupyter pandas requests openpyxl`. This is less clean but will work fine for learning.

## 0.6b Why run order matters

A notebook keeps values in memory while the session is running.

That means:
- If a variable is created in an earlier cell, later cells can reuse it.
- If you run a later cell first, it may fail because the variable does not exist yet.
- Restarting the kernel clears memory, so you need to run cells again from the top.

Rule of thumb: for first runs, execute top-to-bottom.

````python
# Demo step 1: run this cell first
city = "Amsterdam"
print(f"Saved city in memory: {city}")
````

## 0.8 Sneak peek: what you'll be able to build

Run the cell below. **You don't need to understand it yet!** But you will by the end of this course. This is just to show you what's possible in the wonderfull world of API's.

````python
# Don't worry about understanding this yet!
# By the end of this course, you'll know exactly how every line works.

import requests                              # A library for talking to APIs
import pandas as pd                          # A library for working with data tables

# Step 1: Ask an API for data about European countries
response = requests.get(
    "https://restcountries.com/v3.1/region/europe",
    params={"fields": "name,capital,population,area"}
)

# Step 2: Convert the response into a list we can work with
countries = response.json()

# Step 3: Build a clean table from the data
rows = []
for country in countries:
    rows.append({
        "Country": country["name"]["common"],
        "Capital": country["capital"][0] if country.get("capital") else "N/A",
        "Population": country["population"],
        "Area (km²)": country.get("area", 0)
    })

df = pd.DataFrame(rows)
df = df.sort_values("Population", ascending=False)

# Step 4: Show the top 10
print("Top 10 European countries by population:")
print()
print(df.head(10).to_string(index=False))

# Step 5: Save to a file
df.to_csv("european_countries_preview.csv", index=False)
print(f"\nSaved {len(df)} countries to european_countries_preview.csv")
````

If that worked, you just:
1. Called a real API on the internet
2. Processed structured data
3. Created a sorted table
4. Saved it as a CSV file

## 0.10 Troubleshooting

| Problem | Solution |
|---------|----------|
| `python3` not found | Try `python` or 'python3.11' instead. If neither works, reinstall from Self Service. |
| `pip` not found | Make sure virtual environment is activated (you should see `(venv)`) |
| `jupyter` not found | Make sure you ran `pip install jupyter` with venv activated |
| Browser doesn't open | Copy the localhost URL from terminal into your browser |
| Kernel won't start | Restart terminal, reactivate venv, try again |
| PowerShell won't run activate script | Run `Set-ExecutionPolicy -ExecutionPolicy RemoteSigned -Scope CurrentUser` first |
| Proxy/SSL errors | See section 0.5 |

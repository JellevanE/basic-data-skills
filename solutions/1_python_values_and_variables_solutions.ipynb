## 1.1 Variables: giving names to things

Variables store values so you can use them later. The name goes on the left, the value on the right.

```python
country_name = "Netherlands"
population = 17000000
is_in_europe = True

print(country_name)
print(population)
print(is_in_europe)
```

## 1.2 Text (strings)

Strings are text, wrapped in quotes. You can use either single or double quotes.

```python
country = "Netherlands"
capital = 'Amsterdam'    # Single or double quotes both work

# Useful operations
print(len(country))              # 11 (number of characters)
print(country.lower())           # "netherlands"
print(country.upper())           # "NETHERLANDS"
print("land" in country)         # True (checks if "land" appears in the text)
```

## 1.3 Numbers

Python has two types of numbers: **integers** (whole numbers) and **floats** (decimals). Most of the time you don't need to worry about the difference.

```python
population = 17000000      # Integer (whole number)
latitude = 52.37           # Float (decimal number)

# Basic math
print(population / 1000000)       # 17.0
print(latitude + 1)               # 53.37
print(round(52.3789, 2))          # 52.38
```

**Exercise:** The Netherlands has a population of 17,000,000 and an area of 41,543 km². Calculate the population density (population divided by area). Round it to 1 decimal.

```python
population = 17000000
area = 41543
density = round(population / area, 1)
print(density)  # 409.2
```

## 1.4 Booleans

Booleans are `True` or `False`. They're used for yes/no decisions.

## 1.5 Checking types with `type()`

When something unexpected happens, `type()` helps you understand what you're looking at.

```python
print(type("Netherlands"))   # <class 'str'>
print(type(17000000))        # <class 'int'>
print(type(52.37))           # <class 'float'>
print(type(True))            # <class 'bool'>
```

This becomes very useful when debugging API responses — "is this a number or a string that looks like a number?"

## 1.6 f-strings: building dynamic text

f-strings let you insert variables into text. Put `f` before the opening quote and variables inside `{}`.

```python
country = "Netherlands"
population = 17000000

print(f"The population of {country} is {population}")

# You can also do calculations inside the braces
print(f"That's {population / 1000000} million people")
```

You'll use f-strings constantly — for printing results, building messages, and formatting output.

## 1.7 Imports: using other people's code

Python has a huge ecosystem of **libraries** — code that other people wrote so you don't have to. You load them with `import`.

```python
import math
print(math.sqrt(144))    # 12.0 — square root

import datetime
print(datetime.date.today())  # Today's date
```

Later in this course, you'll use:
- `import requests` — for calling APIs
- `import pandas` — for working with data tables
- `import json` — for reading/writing JSON files

The pattern is always the same: `import library_name`, then use it with `library_name.something()`.
